```python
from selenium.webdriver.common.by import By   #2 main
from selenium.common.exceptions import StaleElementReferenceException
import pandas as pd
from selenium.webdriver.common.keys import Keys
import os
import time
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
import random
import re
# Replace with the actual URL
base_url = "https://myneta.info/Bihar2020/index.php"
action_url = "show_candidates"
details_dict = {}

# how many time chrome has to scrool down in the page to get the full list as the page is dynamicaly loaded
no_of_pagedowns = 1

# Define a list of user-agent strings to simulate different browsersno
user_agents = [
    'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.96 Safari/537.36',
    'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Firefox/88.0.4324.96 Safari/537.36',
    'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Edge/88.0.4324.96 Safari/537.36'
]


# Choose a random user-agent string from the list
headers = {
    'User-Agent': random.choice(user_agents)
}

try:
    Content = []
    for constituency_id in range(1,5):
        user_agent = random.choice(user_agents)

        # Set user-agent for Selenium WebDriver
        chrome_options = webdriver.ChromeOptions()
        chrome_options.add_argument(f'user-agent={user_agent}')

        # Create a Chrome WebDriver instance with the specified options
        browser = webdriver.Chrome(options=chrome_options)

        url = f"{base_url}?action={action_url}&constituency_id={constituency_id}"
        # Open the URL in the browser
        browser.get(url)

        # gets the body tag from the html under which content is present
        elem = browser.find_element(By.TAG_NAME, "body")

        while no_of_pagedowns:                      # scrolls the pages to the numeber of times
            elem.send_keys(Keys.PAGE_DOWN)          # pressed the down key
            time.sleep(0.4)                         # waits for the page to load
            no_of_pagedowns -= 1
        #no_of_pagedowns = 1

        html = browser.page_source              # gets the page source once it's loaded
        # extracts the features using bs4
        soup = BeautifulSoup(html, features="html.parser")
        # print("******Closing****")
        # browser.close()                         # closes the broswer ( important !!)
        # Construct the URL for the specific constituency ID

        # Set default values for constituency and district
        constituency = "Unknown"
        district = "Unknown"     #- -
        state = "Bihar"
        
        time.sleep(0.4)
        # Find the div element
        div_element = soup.find('div', class_='w3-panel w3-leftbar w3-light-gray')

        # Extract state and constituency information
        #state_a = div_element.find_all('a')[2]
        #constituency_b = div_element.find('b')
        district_a = div_element.find_all('a')[2]
        constituency_b = div_element.find('b')
        print(district_a,constituency_b)

        if district_a and constituency_b:  
            district = district_a.get_text().strip()
            constituency = constituency_b.get_text().strip()
            

        '''if state_a and constituency_b:
            state = state_a.get_text().strip()
            constituency = constituency_b.get_text().strip()*/'''

        table = soup.find('table', class_='w3-table w3-bordered')
        # Extract the <tbody> content
        tbody = table.find('tbody')

        # Iterate through rows (skipping the header row)
        constituency_data = []
        for row in tbody.find_all('tr')[1:]:
            row_data = []
            # Extract data from cells
            cells = row.find_all(['td', 'th'])
            candidate_name = cells[0].find('a').get_text(strip=True)
            cand_image_name = candidate_name.split('(')[0].strip()
            party = cells[1].get_text(strip=True)
            criminal_cases = cells[2].get_text(strip=True)
            education = cells[3].get_text(strip=True)
            age = cells[4].get_text(strip=True)
            # total_assets = cells[5].get_text(strip=True)
            # liabilities = cells[6].get_text(strip=True)
            # Extract the formatted value using regex for Total Assets
            total_assets_raw = cells[5].get_text(strip=True)
            total_assets_match = re.search(
                r'(~\s*\d+[\d,]*\s*[A-Za-z]+)', total_assets_raw)
            total_assets = total_assets_match.group(
                0) if total_assets_match else ""

            liabilities_raw = cells[6].get_text(strip=True)
            liabilities_match = re.search(
                r'(~\s*\d+[\d,]*\s*[A-Za-z]+)', liabilities_raw)
            liabilities = liabilities_match.group(0) if liabilities_match else "0 Lacs"

            if party in ['BJP', 'INC', 'BSP', 'RJD', 'LJP'] or 'Winner' in cells[0].get_text(strip=True):
                row_data.append(candidate_name + ' (Age ' + age + ' yrs)')
                row_data.append(
                    party + " ( " + constituency + ", " + state + ", MP )")
                row_data.append("Education: " + education)
                row_data.append("Total Assets: " + total_assets)
                row_data.append("Total Liabilities: " + liabilities)
                row_data.append("Criminal Cases: " + criminal_cases)
                row_data.append(constituency)
                row_data.append(district)
                row_data.append(state)

                unique_name = candidate_name.split('(')[0].strip().replace(' ', '_').lower()
                unique_key = unique_name + "_" + constituency.strip().replace(' ', '_').lower() + "_" + district.strip().replace(' ', '_').lower()
                row_data.append(unique_key)
                
                constituency_data.append(row_data)               
               
                party_des = {
                        'BJP': "Bharatiya Janata Party",
                        'INC': "Indian National Congress",
                        'BSP': "Bahujan Samaj Party",
                        'RJD': "Rashtriya Janata Dal",
                        'LJP': "Lok Jan Shakti Party",
                        'JD(U)': "Janata Dal (United)" 
                    }
                
                              
                try:
                    eci_image(cand_image_name, unique_key, party_des.get(party, "Winner"))
                    time.sleep(0.1)
                
                except StaleElementReferenceException:
                    #eci_image(cand_image_name, unique_key, party_des(party))
                    #time.sleep(0.1)
                    pass
                                       
                except Exception as e:
                    print(f"An unexpected error occurred: {e}")
                    eci_image(cand_image_name, unique_key, party_des.get(party))
                    time.sleep(0.1)
                
                    
                    #pass# Break out of the loop for other exceptions
            else:    #- -    
                pass
                  
        Content.extend(constituency_data)  #- -
        
    print(Content)
            
    
    #print(row_data)
    df = pd.DataFrame(Content, columns=['Candidate Name', 'Party', 'Education', 'Total Assets', 'Total Liabilities', 'Criminal Cases','Constituency','District','State','unique key' ])
    df.to_csv(r'C:\Users\lenovo\Desktop\bihar_leaders' + 'mla-bihar.csv',index=None,encoding='utf-8')

except Exception as e:
    df = pd.DataFrame(Content, columns=['Candidate Name', 'Party', 'Education', 'Total Assets', 'Total Liabilities', 'Criminal Cases','Constituency','District','State','unique key' ])
    df.to_csv(r'C:\Users\lenovo\Desktop\bihar_leaders' + 'mla-bihar.csv',index=None,encoding='utf-8')

 
```
